# Dataset Analysis : TMDb Database

## Indroduction

This notebook focuses Analysis and Investigation The Movie Database (TMDb) movie dataset .
The dataset focuses on movies and it's cast, genre and how much revenue it made in box office. The Dataset contains 10866 records on movies. The goal is to focus on the relationship budget spent on production and revenue generated from movie release.

```python
import pandas as pd

df = pd.read_csv("tmdb-movies.csv")
df.sort_values(["popularity"], ascending=0)
```

```python
df.info()
```

The Dataset selects out 1000 records for analysis sorted based on profit from revenue

```python
df = df.sort_values(["budget", "revenue"],ascending=[0,0])
df["profit"] = df["revenue"] - df["budget"]
workframe = df[["original_title", "profit", "revenue", "budget"]].sort_values(["profit"], ascending=0).head(1000)
workframe
```

The Variables we focus here is revenue, budget and profit. Revenue and Budget are indipendent parameters whereas Profit is dependant on budget and profit. Since the three variables are non-null and available for all records there not require cleanning

## Analysis

The analysis done on basic statistics selected 1000 records

```python
workframe.describe()
```

### Maximum budget spent on Movie Production:

```python
workframe[["original_title","budget"]][workframe["budget"] == workframe["budget"].max()]
```

### Maximum revenue generated movie

```python
workframe[["original_title","revenue"]][workframe["revenue"] == workframe["revenue"].max()]
```

### Maximum profit generated movie

```python
workframe[["original_title","profit"]][workframe["profit"] == workframe["profit"].max()]
```
