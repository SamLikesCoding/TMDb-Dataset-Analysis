## Question

The question is here to find out movie proudctions of high gross profit and low budget. The goal is to find trend in
budget expediture in movie production. The Dataset selects out 1000 records for analysis sorted based on profit from revenue

```python
df = df.sort_values(["budget", "revenue"],ascending=[0,0])
df["profit"] = df["revenue"] - df["budget"]
workframe = df[["original_title", "profit", "revenue", "budget"]].sort_values(["profit"], ascending=0).head(1000)
#nz_mean = workframe[workframe["budget"] != 0].mean()
#workframe.fillna(nz_mean, inplace=True)
workframe
```

The Variables we focus here is revenue, budget and profit. Revenue and Budget are indipendent parameters whereas Profit is dependant on budget and profit. Since the three variables are non-null and available for all records there not require cleanning
